```python
import pandas as pd

df = pd.read_csv('./database/jobs-and-salaries-in-data-science/jobs_in_data.csv')

print(df.head())
print(df.columns)
print(df.size)
```

```python
work_year_unique = df['work_year'].unique()
print(work_year_unique)

print('')

for year in df['work_year'].unique():
    print(year)

print('')

work_year_unique_df = pd.DataFrame({'work_year_unique':df['work_year'].unique()})
print(work_year_unique_df)
```

```python
soma_salario_por_ano = df.groupby('work_year')['salary'].sum().reset_index()

print(soma_salario_por_ano)
```

```python
import matplotlib.pyplot as plt

plt.plot(soma_salario_por_ano['work_year'], soma_salario_por_ano['salary'])
plt.show()
plt.bar(soma_salario_por_ano['work_year'], soma_salario_por_ano['salary'])
plt.show()

plt.figure(figsize=(10,6))
plt.plot(
    soma_salario_por_ano['work_year'], soma_salario_por_ano['salary'],
    color='green',
    linewidth=5,
    linestyle='dashed',
    marker='s',
    markersize=10,
    markerfacecolor='black',
    markeredgecolor='blue',
)
plt.xlabel('work_year')
plt.ylabel('salary (10^9)')
plt.show()
```

```python
media_salario_por_ano = df.groupby('work_year')['salary'].mean().round(2).reset_index()

print(media_salario_por_ano)

ax1 = plt.subplot()
ax2 = ax1.twinx()

ax1.plot(media_salario_por_ano['work_year'], media_salario_por_ano['salary'])
ax2.bar(soma_salario_por_ano['work_year'], soma_salario_por_ano['salary'])
plt.show()
```
