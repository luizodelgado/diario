```python
# reforçando o que foi aprendido - importar dados e imprimir informações sobre o df

import pandas as pd

df = pd.read_csv('./database/jobs-and-salaries-in-data-science/jobs_in_data.csv')

colunas = pd.DataFrame({'colunas':df.columns})

print(colunas)
print()
print(df)
```

```python
# reforçando o que foi aprendido - listagem de valores distintos

work_year_unique_array = df['work_year'].unique()
print(work_year_unique_array)

print()

for year in df['work_year'].unique():
    print(year)

print()

work_year_unique_df = pd.DataFrame({'work_year_unique_df':df['work_year'].unique()})
print(work_year_unique_df)
```

```python
# reforçando o que foi aprendido - funções de agregação

soma_salario_ano = df.groupby('work_year')['salary'].sum()
print(soma_salario_ano)

print()

soma_salario_ano_df = df.groupby('work_year')['salary'].sum().reset_index()
print(soma_salario_ano_df)


print()

media_salario_ano_df = df.groupby('work_year')['salary'].mean().reset_index()
print(media_salario_ano_df)
```

```python
# reforçando o que foi aprendido - gráficos

import matplotlib.pyplot as plt

plt.plot(soma_salario_ano_df['work_year'], soma_salario_ano_df['salary'])
plt.show()

plt.bar(
    media_salario_ano_df['work_year'], media_salario_ano_df['salary']
    , color = 'pink'
    )
plt.xlabel('work_year')
plt.ylabel('salary')
plt.show()
```

```python
# reforçando o que foi aprendido - gráficos

ax1 = plt.subplot()
ax2 = ax1.twinx()

ax1.plot(soma_salario_ano_df['work_year'], soma_salario_ano_df['salary'])
ax2.bar(media_salario_ano_df['work_year'], media_salario_ano_df['salary'])
plt.show()
```
