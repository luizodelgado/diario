```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../databases/jobs-and-salaries-in-data-science/jobs_in_data.csv')

df.info()

display(df.describe())
```

```python
mean_salary_year = df.groupby('work_year')['salary'].mean().reset_index().round(2)

display(mean_salary_year)
```

```python
max_salary_year = df.groupby('work_year')['salary'].max().reset_index()

display(max_salary_year)
```

```python
agregacoes = pd.DataFrame({
    # 'work_year': df['work_year'].unique(),
    'salary': df.groupby('work_year')['salary'].sum().round(2),
    'mean': df.groupby('work_year')['salary'].mean().round(2),
    'max': df.groupby('work_year')['salary'].max().round(2)
    }
    )

display(agregacoes)
```

```python
agregacoes2 = df.groupby('work_year')['salary'].agg(
    sum='sum',
    mean='mean',
    max='max'
).round(2).reset_index()

display(agregacoes2)
```

```python
agregacoes3 = df.groupby('work_year').agg(
    sum_salary=('salary','sum'),
    mean_salary=('salary', 'mean'),
    max_salary=('salary', 'max'),
    sum_salary_usd=('salary_in_usd', 'sum'),
    mean_salary_usd=('salary_in_usd', 'mean'),
    max_salary_usd=('salary_in_usd', 'max')
).round(2).reset_index()

display(agregacoes3)
```
